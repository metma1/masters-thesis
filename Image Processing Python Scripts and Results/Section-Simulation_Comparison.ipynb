```python
import pandas as pd
import numpy as np
import tkinter
from tkinter import filedialog
import os
import seaborn as sns
import matplotlib.pyplot as plt

sns.color_palette("bright")

sim_pixel_to_micron = 23.20267324
sec_pixel_to_micron = 1.6124
```

# Y55 - S1 Width/Height Comparison

## Width Comparison

```python
file_path = tkinter.filedialog.askopenfile(mode='r')
Y55_S1_Widht_Height = pd.read_csv(file_path, delimiter=";")

fig, ax = plt.subplots(figsize=(20, 8))
plot = sns.lineplot(x='X', y='value', hue='Source', legend='full', data=pd.melt(Y55_S1_Widht_Height, id_vars=['X', 'Source'], value_vars=['Width']), errorbar='sd', err_style='band')
plot.set(xlabel='Days', ylabel='Microns', title='Width of Y55 colony sections and S1 model simulation results')
plot.legend(title='')
sns.move_legend(plot, "upper left")
plt.grid()
```

## Height comparison

```python
fig, ax = plt.subplots(figsize=(20, 8))
plot = sns.lineplot(x='X', y='value', hue='Source', legend='full', data=pd.melt(Y55_S1_Widht_Height, id_vars=['X', 'Source'], value_vars=['Height']), errorbar='sd', err_style='band')
plot.set(xlabel='Days', ylabel='Microns', title='Height of Y55 colony sections and S1 model simulation results')
plot.legend(title='')
sns.move_legend(plot, "upper left")
plt.grid()
```

# Mixed - TwoStrain Strain Ratio Comparison
